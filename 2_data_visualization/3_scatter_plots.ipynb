# Scatter Plots
We've practiced using Seaborn to make line graphs, bar charts, and heatmaps. So nooow, let's do some fun scatter plots.

### Initial Setup
Let's get all our packages imported.

````python
from termcolor import colored

import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
print(colored("Seaborn Setup Complete!", 'green'))
````

### Get Data Imported
We'll be working with a dataset of insurance charges. We'll use this data to understand why some customers pay more than others (hey look, data science!).

````python
# Import data
import my_modules.data_imports as data

# Note that this data doesn't have an index set
insurance_charges = data.import_insurance_data()

insurance_charges.head()
````

### Scatter Plots Galore
Creating a basic scatterplot is as easy as using the ```scatterplot``` function.
For this example we'll use the following data:
- x-axis: BMI (```x=insurance_charges['bmi']```)
- y-axis: Charges (```y=insurance_charges['charges']```)

````python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=insurance_charges['bmi'], y=insurance_charges['charges'])
````

Would you look at that? Pretty positive correlation, customers with higher BMI pay more for insurance! Let's now throw a regression line in there using ```sns.regplot```

````python
plt.figure(figsize=(10,5))
sns.regplot(x=insurance_charges['bmi'], y=insurance_charges['charges'])
````

### Color-coded scatter plots
Love scatter plots but wanted to use more than 2 variables? Then lets add some color!! Now let's see how BMI compares to insurance costs for people who are smokers or not.

````python
plt.figure(figsize=(10,5))
sns.scatterplot(x=insurance_charges['bmi'], y=insurance_charges['charges'], hue=insurance_charges['smoker'])
````

Now isn't that interesting... While BMI does have a small effect on insurance costs, smokers pay tend to pay MUCH more. Let's use a regression line again corresponding to smokers and nonsmokers (```sns.lmplot```)

````python
# lmplot doesn't seem to use matplotlib's plt here, so we need to set height and aspect
sns.lmplot(x="bmi", y="charges", hue="smoker", data=insurance_charges, height=5, aspect=2)
````

### Categorical Scatter Plot
For our last example, we'll use a less common scatter plot known as a **categorical scatter plot**. In this use-case, we will use smoker as our x-axis value instead of BMI (we're back to only using two variables)

````python
plt.figure(figsize=(10,5))
sns.swarmplot(x=insurance_charges['smoker'], y=insurance_charges['charges'])
````
