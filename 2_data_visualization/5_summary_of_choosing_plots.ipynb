# How to Choose Plot Types

So far we've learned about a lot of different plot types and how to use Seaborn to make them. There's many different ways to display data, and some are better than others for various situations. Let's recap the types of plots, broken up into three categories: **Trends**, **Relationships**, and **Distributions**.
Examples of all these plots can be found in the previous notebooks.

<img src=""></img>

- **Trends** - defined as a pattern of change
  - sns.lineplot - **Line Charts** are best for showing trends over a period of time, with the ability to add multiple lines to compare trends of different groups.
- **Relationships** - help to understand relationships between variables
  - sns.barplot - **Bar Charts** are useful for comparing quantities corresponding to different groups.
  - sns.heatmap - **Heatmaps** can be used to find color-coded patterns in tables of numbers.
  - sns.scratterplot - **Scatter plots** help show the relationships between two continuous variables, and when color-coded can show the relationship with a third categorical variable.
  - sns.regplot - **Regression Lines** are good for using in your scatterplot to help see linear relationships between variables. 
  - sns.lmplot - **Multiple Regression Lines** are great for scatterplots with multiple color-coded groups.
  - sns.swarmplot - **Categorical Scatter Plots** show relationships between a continious variable and a categorical variable.
- **Distributions** - used to visualize the possible values that we can expect to see in a variable, and how likely they are.
  - sns.displot - **Histograms** show the distribution of a single numerical variable
  - sns.kdeplot - **KDE plots (2D kernel density estimates)** show an estimated, smooth distribution of a single (or two) numerical variable(s)
  - sns.jointplot - Displays 2D KDE plot with the corresponding KDE plots for each individual variables.
