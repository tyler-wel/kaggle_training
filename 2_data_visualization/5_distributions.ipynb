# Distributions
Let's get into some more fun, "data-y" graphs with histograms and density plots.

### Initial Setup
First off, get all the packages imported as always.


````python
from termcolor import colored

import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
print(colored("Seaborn Setup Complete!", 'green'))
````

### Import Data
For this example, we'll be working with a dataset of 150 different flowers, 50 each from three different species (*Iris Setosa*, *Iris Versicolor*, *Iris Virginica*)

````python
# Import data
import my_modules.data_imports as data

# Note that this data doesn't have an index set
flower_data = data.import_flower_data()

flower_data.head()
````

### Histograms
Let's start off with a simple histogram to see how the petal length varies between the different iris flowers using ```sns.distplot```.

#### Note: 
The kde variable is an optional boolean parameter for whether or not we want a guassian kernel density estimate.

````python
plt.figure(figsize=(10,4))
sns.distplot(a=flower_data['Petal Length (cm)'], kde=False)
````

### Density Plots
Next, let's make some kernel density estimates! Despite our note on the histogram above, we won't be using the ```sns.distplot``` function, but instead the ```sns.kdeplot``` function. KDE's Are a kind of smoothed out histogram to give us a better picture at values in between our provided values.

````python
sns.kdeplot(data=flower_data['Petal Length (cm)'], shade=True)
````

### 2D KDE Plots
One dimensional plots are fun and all, but what about the amazing second dimension. These graphs give us a sort of "height". We use the ```sns.jointplot``` function to achieve this.

In the plot below we'll be comparing Petal Length to Sepal Width. The darker parks will show us the likelihood of the different combinations of lengths.

````python
sns.jointplot(x=flower_data['Petal Length (cm)'], y=flower_data['Sepal Width (cm)'], kind="kde")
````

### Color-coded plots

For the next bit, we'll go about making color-coded plots using different data sets to compare things. We'll take the same flower data, but seperated into three different sets for each species. So let's get the imported first.

````python
# All the flowers!
iris_set = data.import_iris_set()
iris_ver = data.import_iris_ver()
iris_vir = data.import_iris_vir()

# Let's just inspect one
iris_set.head()
````

With the data imported, let's go make a different histogram for each of the species using our previous ```sns.distplot``` function. We can use the ```label=``` parameter to add it to the legend so we know what color is which. 

````python
plt.figure(figsize=(10,5))
plt.title("Histogram of Petal Lengths, by Species")

sns.distplot(a=iris_set['Petal Length (cm)'], label="Iris-setosa", kde=False)
sns.distplot(a=iris_ver['Petal Length (cm)'], label="Iris-versicolor", kde=False)
sns.distplot(a=iris_vir['Petal Length (cm)'], label="Iris-virginica", kde=False)

# Toggle to show the legend
plt.legend()
````

And just for some extra practice, let's do the same thing with the KDE plot!

````python
plt.figure(figsize=(10,5))
plt.title("Distribution of Petal Lengths, by Species")

sns.kdeplot(data=iris_set['Petal Length (cm)'], label="Iris-setosa", shade=True)
sns.kdeplot(data=iris_ver['Petal Length (cm)'], label="Iris-versicolor", shade=True)
sns.kdeplot(data=iris_vir['Petal Length (cm)'], label="Iris-virginica", shade=True)

plt.legend()
````
