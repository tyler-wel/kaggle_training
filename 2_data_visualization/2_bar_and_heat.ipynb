# Bar Charts & Heatmaps!
In this notebook, we'll be learning more about data graphing and Seaborn options with bar charts and heatmaps!

### Initial Setup
As before, import required libraries.

````python
from termcolor import colored

import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
print(colored("Seaborn Setup Complete!", 'green'))
````

### Data Import
Before importing the data, it's always a good idea to open it up and take a look at what kind of things we'll be working with. Mainly, which column we want to use as our index, as well as the need to parse any dates. 
In this case, we'll be taking a look at the average arrival delays of American Airlines flights. Our index will be the first column, 'Months'

````python
# Import data
import my_modules.data_imports as data
flight_delays = data.import_flight_delays()

# Peek the data
flight_delays
````

### Creating Bar Charts
For our first example, let's graph the average arrival delay for Spirit Airlines (code: NK) by the month.

#### Note:
When plotting barcharts, you must use ```data.index``` for the x value instead of ```data['Column']```

````python
# Set width/height
plt.figure(figsize=(10,6))

# Add title for fun
plt.title("Avg Arrival Delay for Spirit Airline Flights (by month)")

# Simple bar chart
sns.barplot(x=flight_delays.index, y=flight_delays['NK'])

# Vert axis label
plt.ylabel("Arrival Delay (min)")
````

### Heatmaps
Now let's do the same thing as before, but with a heatmap! Heatmaps break up our data by color, showing different levels of 'intensity'. We will see shortly why this works pretty well with our data set.

````python
# Set Width/Height
plt.figure(figsize=(14, 7))

# Nice title
plt.title("Avg Arrival Delay for Each Airline, by Month")

# Our wonderful heatmap (annot ~ annotate the cells with the values)
sns.heatmap(data=flight_delays, annot=True)

# Label for niceness
plt.xlabel("Airline")

# Bandaid fix for heatmap bug until matplotlib updates to 3.1.2
plt.ylim(len(flight_delays)-0.5, -0.5)
````
