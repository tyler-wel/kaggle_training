# Learning with Line Charts!
In this notebook, we'll be exploring data analysis and creating charts with line charts.

### Initial Setup
Let's get our pandas, matlib, and seaborn imported


````python
from termcolor import colored

import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
print(colored("Seaborn Setup Complete!", 'green'))
````

### Plotting data
Now that we got a look at the data, let's get to plotting it! With Seaborn, it is as easy as one simple line of code! 

````python
sns.lineplot(data=spotify_data)
````

### Customizing Plots
One line of code and a working graph! But it's kind of cramped, and there's times where we will want to customize it more. So let's get to adding some addition parameters.

````python
# Remember, matlibplot was imported as plt, seaborn imported as sns
# Set width and height
plt.figure(figsize=(14,6))

# Add graph title
plt.title("Daily Global Streams of Popular Songs 2017-2018")

# Line chart, go!
sns.lineplot(data=spotify_data)
````

Looking much more professional!

### Plotting Subset Data
Time to get even more customizable by graphing subsets of our dataset. First we'll take a look at what columns we have to work with, then use these columns to graph specific data.

````python
list(spotify_data.columns)
````

Let's go ahead and compare daily streams of 'Shape of You' vs 'Despacito'

````python
# Set the size of our graph
plt.figure(figsize=(14, 6))

# Set the title so we know wtf we're lookin at
plt.title("Daily Global Streams of 'Shape of You' vs. 'Despacito")

# Line chart showing 'Shape of You'
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")

# Line chart showing 'Despacito'
sns.lineplot(data=spotify_data['Despacito'], label="Despacito")

# Add X-Axis label
plt.xlabel("Date")
````

Now it's nice and easy to compare two different pieces of data. Notice in previous examples, a single ```lineplot``` statement plotted all the data. In the above case, we were able to use the ```lineplot``` function twice, and instead of plotting two graphs, it plots the two lines on the same graph!
