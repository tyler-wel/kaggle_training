````python
from termcolor import colored

import pandas as pd
pd.plotting.register_matplotlib_converters()

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
print(colored("Seaborn Setup Complete!", 'green'))

import os
filepath = ("./datasets/data-vis/fifa.csv")
if os.path.exists(filepath):
    print(colored("File exists, ready to go!", 'green'))
````

For this introduction, we'll be using a FIFA data set to plot. Import it below!

Note:
- ```index_col``` param is generally the name of the first column, used to index our data

````python
fifa_data = pd.read_csv(filepath, index_col="Date", parse_dates=True)
````

With the data loaded, time to plot our first line chart! Seaborn provides high-level interface drawing nice plots and is based off of matplotlib. 


````python
# Use matplotlib to set the width and height of the figure (size in inches)
plt.figure(figsize=(16, 6))

# Use seaborn for line chart!
sns.lineplot(data=fifa_data)
````
